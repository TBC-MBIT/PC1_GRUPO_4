```python
# Carga los archivos CSV en DataFrames de pandas
df_tinto = pd.read_csv("winequality-red.csv", sep=";")
df_blanco = pd.read_csv("winequality-white.csv", sep=";")
```

```python
# Creo la lista con las columnas numéricas.
columnas = df_vinos.select_dtypes(include=["int64", "float64"]).columns
```

```python
# Compruebo que tengo las columnas correctas
columnas
```

```python
# Compruebo el número de filas del dataset antes de borrar valores nulos
df_vinos.shape
```

```python
# Eliminación de valores ausentes en columnas numéricas
df_vinos_limpia = df_vinos.dropna(subset=columnas)
```

```python
# Comprobamos que no existían valores ausentes. Siguen saliendo el mismo número de filas y columnas.
df_vinos_limpia.shape
```

```python
# Creo dos variables para recoger el valor de la columna numerica en cuartil 25 % y cuartil 75% y una tercera variable
# con la diferencia entre ambas, para obtener el rango típico.
q1 = df_vinos[columnas].quantile(0.25)
q3 = df_vinos[columnas].quantile(0.75)
IQR = q3-q1
```

```python
# Creo una función para que devuelva solo los valores atípicos de cada columna 
def detect_atipicos(series):
    inferiores = q1.loc[series.name] - 1.5 * IQR.loc[series.name]
    superiores = q3.loc[series.name] + 1.5 * IQR.loc[series.name]
    return series[(series < inferiores) | (series > superiores)]
```

```python
# Creo un diccionario con los valores atípicos
atipicos = {}
for col in columnas:
    atipicos[col] = detect_atipicos(df_vinos_limpia[col])
```

```python
#Compruebo que los índices obtenidos son válidos y existen dentro del Dataframe y elimino las coincidencias con atípicos
for col in atipicos:
    valid_index = df_vinos_limpia.index.intersection(atipicos[col].index)
    df_vinos_limpia = df_vinos_limpia.drop(valid_index)
```
