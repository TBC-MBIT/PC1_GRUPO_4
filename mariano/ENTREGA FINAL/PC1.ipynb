```python
Creamos la columna de si es vino tinto o blanco en base al indicador de realizar la combinacion(merge).
Para ello, mapeamos la columna _merge y si pertenece a la derecha es un vino blanco y si pertence a la izquierda es un vino tinto. En caso de que 
pertenezca a los dos, lo denominamos Rosado ya que es el otro tipo de vino que puede existir en base al color del vino.
Creamos una función para detectar el tipo de vino.
```

```python
Se procede a realizar el filtrado de valores atípicos a nivel estadístico, para ello, se considera como valores atípicos a los valores que se
encuentren por debajo del percentil 25% o por encima del percentil 75% en un rango superior o inferior a 1.5
```

```python
Se crean dos variables para los valores del 25 y 75%, y otra variable para obtener el rango típico(IQR). Se guarda en el dataframe df_percentil para 
una mejor visualizacion de los resultados obtenidos al obtener los valores percentiles y así comparar con el dataframe original y filtrar los datos.
```

```python
columns = ['Fix_acidity','vol_acidity','citric_acid','res_sugar','Chlorides','free_sulfurDioxide','total_sulfurDioxide','Density','PH','Sulphates','Alcohol','Quality']
Q1=df_wines[columns].quantile(.25)
Q3=df_wines[columns].quantile(.75)
IQR=Q3-Q1

df_estadistico = pd.DataFrame(Q1)
df_estadistico['Q1'] = Q1
df_estadistico['Q3'] = Q3
df_estadistico['IQR'] = IQR
df_estadistico
df_percentil = df_estadistico[['Q1','Q3','IQR']]
df_percentil['Inferior'] = df_percentil['Q1'] - 1.5 * df_percentil['IQR']#Realizamos la operación para obtener el valor inferior y lo metemos en el dataFrame
df_percentil['Superior'] = df_percentil['Q3'] + 1.5 * df_percentil['IQR']#Realizamos la operación para obtener el valor superior y lo metemos en el dataFrame
df_percentil.reset_index()
```

```python
def detect_atipicos(series):

    inferiores = Q1.loc[series.name] - 1.5 * IQR.loc[series.name]

    superiores = Q3.loc[series.name] + 1.5 * IQR.loc[series.name]

    return series[(series < inferiores) | (series > superiores)]
```

```python
# Creo un diccionario con los valores atípicos
atipicos = {}
#atipicos
df_limpio = df_wines
df_limpio
```

```python
for col in columns:

    atipicos[col] = detect_atipicos(df_limpio[col])
```

```python
Compruebo que los índices obtenidos son válidos y existen dentro del Dataframe
```

```python
for col in atipicos:

    valid_indices = df_limpio.index.intersection(atipicos[col].index)

    df_limpio = df_limpio.drop(valid_indices)

df_limpio
```
